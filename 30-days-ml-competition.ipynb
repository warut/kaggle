```python
# Load the training data
train = pd.read_csv("../input/30-days-of-ml/train.csv", index_col=0)
test = pd.read_csv("../input/30-days-of-ml/test.csv", index_col=0)

# Preview the data
train.head()
```

```python
# Separate target from features
y = train['target']
features = train.drop(['target'], axis=1)

# Preview features
features.head()
```

```python
# List of categorical columns
object_cols = [col for col in features.columns if 'cat' in col]
#object_cols.remove('cat9')
# ordinal-encode categorical columns
X = features.copy()
#X.drop(['cat9','cont4','cont13'],axis=1,inplace=True)
X_test = test.copy()
#X_test.drop(['cat9','cont4','cont13'],axis=1,inplace=True)
ordinal_encoder = OrdinalEncoder()
X[object_cols] = ordinal_encoder.fit_transform(features[object_cols])
X_test[object_cols] = ordinal_encoder.transform(test[object_cols])

# Preview the ordinal-encoded features
X.head()
```

```python
#Preprocessing
from sklearn.preprocessing import StandardScaler,PolynomialFeatures

X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state=0)

#Polynomail
#pmod = PolynomialFeatures(degree=2)
#pmod.fit_transform(X_train,y_train)

#StandardScaler
#pmod = StandardScaler()
#pmod.fit_transform(X_train,y_train)
```

```python
from sklearn.linear_model import RANSACRegressor,SGDRegressor,BayesianRidge,TweedieRegressor,LinearRegression,PassiveAggressiveRegressor
from sklearn.pipeline import make_pipeline

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold

#lightgbm parameter
from xgboost import XGBRegressor

model = XGBRegressor(random_state=1, 
                         n_jobs=-1,
                         n_estimators= 2206,
                         tree_method='gpu_hist',
                         learning_rate= 0.02604083138419779,
                         subsample= 0.7980218163372579,
                         col_samplebytree = 0.13641457979856397,
                         max_depth= 5,
                         reg_alpha = 25.85614181114505, 
                         reg_lambda = 9.260067417192285,
                         booster='gbtree', objective='reg:squarederror',
                        # eval_metric='rmse'
                    )

#kf = KFold(n_splits=5, shuffle=True, random_state=42).split(X_train, y_train)
model.fit(X_train,y_train)

preds_valid = model.predict(X_valid)
print(mean_squared_error(y_valid, preds_valid, squared=False))
```
